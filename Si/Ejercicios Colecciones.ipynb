## Ejercicio 3
Crea una lista nueva con todas las combinaciones de las siguientes dos listas:
> ```ejer_3_1 = ["Hola", "amigo"]```
>
> ```ejer_3_2 = ["Que", "tal"]```

Obten el siguiente output:

['Hola Que', 'Hola tal', 'amigo Que', 'amigo tal']

````python
ejer_3_1 = ["Hola", "amigo"]
ejer_3_2 = ["Que", "tal"]
##Se usa "[f"{x} {y}" for x in ejer... for y un ejer... para hacer más sencillo el codigo 
resultado = [f"{x} {y}" for x in ejer_3_1 for y in ejer_3_2]

print(resultado)
````

Consejo: No intentes eliminar elementos sobre la lista que estás iterando. Haz una copia con `ejer_5.copy()`.

````python
ejer_5 = [3, 20, 3, 47, 19, 3, 29, 45, 67, 78, 90, 3, 3]
ejer_5_copia = ejer_5.copy()

for elemento in ejer_5_copia:
    if elemento == 3:
        ejer_5.remove(3)
        print(ejer_5)
print(ejer_5)
print("Lista original?", ejer_5_copia)
print(ejer_5_copia)
````

## Ejercicio 12
Convierte la siguiente tupla en un diccionario
> ```ejer_12 = [("x", 1), ("x", 2), ("x", 3), ("y", 1), ("y", 2), ("z", 1)]```

TIP: Puedes usar dos variables de control en un bucle `for`. Una por cada valor de las tuplas.

````python
ejer_12 = [("x", 1), ("x", 2), ("x", 3), ("y", 1), ("y", 2), ("z", 1)]

dict_resultado = {}

for clave, valor in ejer_12:
    dict_resultado[clave] = valor

print(dict_resultado)
````
